```python
import sqlite3
%config SqlMagic.style = '_DEPRECATED_DEFAULT'
connection = sqlite3.connect('my_database2.db')
%load_ext sql
%sql sqlite:///my_database2.db
```

```python
%%sql
SELECT Product.name as tovar, Provider.name as postavshik FROM Store
INNER JOIN Product ON Store.ID_Product = Product.id
INNER JOIN Provider ON Store.ID_Provider = Provider.id
```

```python
%%sql
SELECT Product.name, Provider.name, quantity FROM Product
LEFT JOIN Store ON Product.id = Store.ID_Product
LEFT JOIN Provider ON Store.ID_Provider = Provider.id
```

```python
%%sql
SELECT Product.name, Deal.quantity FROM Deal
RIGHT JOIN Store ON Store.id = Deal.ID_Store
RIGHT JOIN Product ON Product.id = Store.ID_Product
```
